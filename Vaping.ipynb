```python
df_2=pd.read_csv('C:/Users/arnav/Downloads/vaportalk_general.csv')
```

```python
vape_general=df_2.dropna(subset=['comment'])[["topics",'comment']]

vape_general.head()
```

```python
import re
# Remove punctuation
vape_general['comment_clean'] = vape_general['comment'].map(lambda x: re.sub('[,\.!?]', '', x))
# Convert the titles to lowercase
vape_general['comment_clean'] = vape_general['comment'].map(lambda x: x.lower())
# Print out the first rows of papers
vape_general['comment_clean'].head()

vape_general['comment_clean']
```

```python
from wordcloud import WordCloud
# Join the different processed titles together.
long_string = ','.join(list(vape_general['comment_clean'].values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, contour_color='steelblue')
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
```

```python
comments=list(vape_general['comment'].drop_duplicates().values)


import spacy

spacy.load('en_core_web_sm')
from spacy.lang.en import English
parser = English()
def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens    
    
```
